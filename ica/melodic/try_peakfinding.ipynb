```python
from os.path import join as pjoin
import os
import numpy as np
import seaborn as sns
import scipy
from scipy.signal import periodogram, find_peaks
import matplotlib.pyplot as plt
```

```python
melodic_outdir_s1 = '/home/homeGlobal/oli/somato/scratch/ica/MELODIC/melodic_wf_workdir/subject_lvl/somato_melodic_wf/melodic/mapflow/_melodic0/'
comp1_tsfile = pjoin(melodic_outdir_s1, 'report/t29.txt')
ts_arr = np.loadtxt(comp1_tsfile)
```

```python
frequencies, powerdensities = periodogram(ts_arr)
```

```python
sns.lineplot(y=ts_arr, x=range(len(ts_arr)))
```

```python
powerstd = np.std(powerdensities)
powermax = np.max(powerdensities)

powerpeaks = find_peaks(powerdensities, prominence=powerstd * 3)
powerpeaks = powerpeaks[0]
```

```python
plt.plot(powerdensities)
plt.plot(powerpeaks, powerdensities[powerpeaks], "x")
```
